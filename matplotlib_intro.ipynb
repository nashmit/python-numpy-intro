### Heatmaps, Contours and Images

```python
x = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, x)

pos = np.dstack((X, Y))

Z = multivariate_normal.pdf(pos, mean=[0, 0])

plt.pcolormesh(X, Y, Z, shading="auto")
```

```python
x = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, x)

pos = np.dstack((X, Y))

Z = multivariate_normal.pdf(pos, mean=[0, 0])

plt.contour(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 1000)

X, Y = np.meshgrid(x, x)

pos = np.dstack((X, Y))

Z1 = multivariate_normal.pdf(pos, mean=[-1, -1])
Z2 = multivariate_normal.pdf(pos, mean=[1, 1])

Z = (Z1 - Z2) > 0

a = multivariate_normal.rvs(size=500, mean=[-1, -1])
b = multivariate_normal.rvs(size=500, mean=[1, 1])

plt.contourf(X, Y, Z, alpha=0.4, colors=["red", "blue"], levels=1)
plt.scatter(a[:, 0], a[:, 1], c="b")
plt.scatter(b[:, 0], b[:, 1], c="r")
```

### Annotations

```python
plt.figure(figsize=(9, 6))
plt.hist(z, bins=50, density=True)
plt.plot(x, norm.pdf(x), linewidth=4)

plt.title("Gaussian Distribution", fontsize=20)
plt.xlabel("x", fontsize=16)
plt.ylabel("$\mathcal{N}(x)$", fontsize=16)


plt.show()
```

```python
plt.figure(figsize=(9, 6))

plt.hist(z, bins=50, density=True, label="Sample Data")
plt.plot(x, norm.pdf(x), linewidth=4, label="Gaussian PDF")

plt.legend(loc=5)

plt.title("Gaussian Distribution", fontsize=20)
plt.xlabel("x", fontsize=16)
plt.ylabel("$\mathcal{N}(x)$", fontsize=16)


plt.show()
```

```python
plt.figure(figsize=(9, 6))

# plt.hist(z, bins=50, density=True, label='Sample Data')
plt.plot(x, norm.pdf(x), linewidth=3, label="Gaussian PDF")

plt.legend()

plt.title("Gaussian Distribution", fontsize=20)
plt.xlabel("x", fontsize=16)
plt.ylabel("$\mathcal{N}(x)$", fontsize=16)

hwhm = np.sqrt(2 * np.log(2)) * norm.std()
plt.axvline(hwhm, c="r")
plt.axvline(-hwhm, c="r")
plt.xticks([-hwhm, hwhm], ["$-\\sqrt{2\ln2}\\sigma$", "$\\sqrt{2\ln2}\\sigma$"])

plt.axhline(norm.pdf(0), c="k", linestyle="--")
plt.axhline(0.5 * norm.pdf(0), c="k", linestyle=":")
plt.yticks([norm.pdf(0), 0.5 * norm.pdf(0)], ["$f_{max}$", "$\\frac{1}{2}f_{max}$"])

plt.show()
```

```python
x = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, x)

pos = np.dstack((X, Y))

Z = multivariate_normal.pdf(pos, mean=[0, 0])

plt.contourf(X, Y, Z)
plt.annotate(
    "center",
    xy=(0, 0),
    xytext=(1, 1),
    arrowprops=dict(facecolor="black", shrink=0.05),
    fontsize=14,
)
plt.show()
```

### Subplots, OO-Interface

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12), squeeze=True, constrained_layout=True)

axes = axes.flatten()

x = np.linspace(-3, 3, 1000)

for i, ax in enumerate(axes):
    ax.plot(x, x ** i)
    ax.set_title(f"Polynomial of Order {i}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

fig.show()
```

```python
fig, axes = plt.subplots(2, 1, sharex=True)

axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
```
